```python
import sqlite3 as sql
import pandas as pd 
import numpy as np 
```

```python
database_connection = sql.connect("../sql_database/Election_Data.db")
db = database_connection.cursor()
```

```python
by_county2016 = pd.read_excel("../data/georgia_elections_data/2016NovemberGeneral/November8_2016(County)/DocumentDirect SSVRZ376R3.xlsx")
```

```python
by_county2016
```

```python
column_names = by_county2016.loc[0].dropna()
column_names = list(column_names)
```

```python
column_names[2:]
["Reg", "Voted", "Percent"]

cols = ["COUNTYNAME", "COUNTYCODE"]
for category in column_names[2:]: 
    for vote_type in ["Reg", "Voted", "Percent"]: 
        cols.append(category.replace(" ", "")+vote_type)
```

```python
by_county2016.rename(columns = {current: new for current, new in zip(by_county2016.columns, cols)}, inplace = True)
```

```python
by_county2016.drop([0, 1], inplace=True)
```

```python
by_county2016.reset_index(inplace = True)
```

```python
by_county2016[cols[2:]] = by_county2016[cols[2:]].astype("float32")
```

```python
by_county2016[cols[1]] = by_county2016[cols[1]].astype('int32')
```

```python
by_county2016[cols[0]] = by_county2016[cols[0]].astype('string')
```

```python
by_county2016.to_sql(name =  "race_gender_by_county_2016", con=database_connection, if_exists='replace')
```
